#### Data Visualization

```python
# visualizing the data as a whole using the pair plot
import seaborn as sns 
sns.pairplot(data)
```

```python
# finding the pairwise correlations between the columns and visualising using heatmaps
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title('Ship Mode')
plt.pie(data['Ship Mode'].value_counts(),labels=data['Ship Mode'].value_counts().index,autopct='%2.2f%%')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.title('Category')
plt.pie(data['Category'].value_counts(), labels=data['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# computing top categories in terms of sales from first 100 observations
category_s = data.groupby("Category").Sales.sum().nlargest(n=100)
# computing top categories in terms of profit from first 100 observations
category_p = data.groupby("Category").Profit.sum().nlargest(n=100)

# plotting to see it visually
plt.style.use('seaborn')
category_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
category_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='RED')
plt.xlabel('Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Category Sales vs Profit",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.title('Sub-Category')
plt.pie(data['Sub-Category'].value_counts(), labels=data['Sub-Category'].value_counts().index)
```

```python
# computing top sub-categories in terms of sales from first 100 observations
subcategory_s = data.groupby("Sub-Category").Sales.sum().nlargest(n = 100)
# computing top sub-categories in terms of profit from first 100 observations
subcategory_p = data.groupby("Sub-Category").Profit.sum().nlargest(n = 100)

# plotting to see it visually
plt.style.use('seaborn')
subcategory_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
subcategory_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Sub-Category Sales vs Profit",fontsize = 15)
plt.show()
```

```python

plt.figure(figsize=(14,12))
statewise = data.groupby(['Sub-Category'])['Profit'].sum().nlargest(50)
statewise.plot.barh() # h for horizontal
```

```python
plt.figure(figsize = (10,10))
plt.title('States')
plt.pie(data['State'].value_counts(), labels=data['State'].value_counts().index)
plt.show()
```

```python
# computing top states in terms of sales from first 10 observations
states_s = data.groupby("State").Sales.sum().nlargest(n=10)

# computing top states in terms of profit from first 10 observations
states_p = data.groupby("State").Profit.sum().nlargest(n = 10)

plt.style.use('seaborn')
states_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
states_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('States',fontsize = 15)
plt.ylabel('Total sales',fontsize = 15)
plt.title("Top 10 states Sales vs Profit",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(14,12))
statewise = data.groupby(['State'])['Profit'].sum().nlargest(50)
statewise.plot.barh() # h for horizontal
```

```python
plt.figure(figsize = (6,6))
plt.title('Region')
plt.pie(data['Region'].value_counts(), labels=data['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
# computing top regions in terms of sales from first 100 observations
top_region_s = data.groupby("Region").Sales.sum().nlargest(n=100)
# computing top regions in terms of profit from first 100 observations
top_region_p = data.groupby("Region").Profit.sum().nlargest(n=100)

# plotting to see it visually
plt.style.use('seaborn')
top_region_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_region_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='red')
plt.xlabel('Region',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Region Sales vs Profit",fontsize = 15)
plt.show()
```

```python
plt.style.use('seaborn')
data.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)

plt.show()
```
